```python

para = ['RBHOM', 'RDFRG', 'RNANA', 'RIODA', 'RCOLI', 'RHIS2', 'RHART', 'RCHIL', 'RGIAR', 'RTRIC']
n = 0
df = pd.DataFrame()
while n < 50000:
    z = []
    t=0
    for i in para:
        r = np.random.randn()
        if r > 0:
            z.append(para[t])
        else:
            z.append('None')

        t=t+1
    n=n+1
    b = ', '.join(z)
    d=[]
    d.append(b)
    df = df.append(d)
df.drop_duplicates(inplace=True)
df.size
df = df.sort([0])
df[0].str[:].str.split(',', expand=True).astype(str)
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
para = ['RBHOM', 'RDFRG', 'RNANA', 'RIODA', 'RCOLI', 'RHIS2', 'RHART', 'RCHIL', 'RGIAR', 'RTRIC']
n = 0
df = pd.DataFrame()
while n < 50000:
    z = []
    t=0
    for i in para:
        r = np.random.randn()
        if r > 0:
            z.append(para[t])
        else:
            z.append('None')

        t=t+1
    n=n+1
    b = ', '.join(z)
    d=[]
    d.append(b)
    df = df.append(d)
df.drop_duplicates(inplace=True)
df.size
df = df.sort([0])
df[0].str[:].str.split(',', expand=True).astype(str)
```

```python
#Import file, summarize date data. 
OrgID = "CGC"
fileName = "Update_Review/Sep/%s_Sep.xls" %OrgID


Org = pd.read_excel(fileName, parse_dates=True)
Org = Org.set_index("Date")
Org = Org.index.value_counts().sort_index()
```

```python
#Generate 22 random dates, calculate length of results for those dates
import random
size = len(Org)
ndf = pd.DataFrame()
r = []
dates = []
count = []
for x in range(22):
    t = random.randint(1,size)
    r.append(t)
for i in r:
    m = Org.index[i]
    n = Org[i]
    dates.append(m)
    count.append(n)
    
ndf['Nums'] = r
ndf['Dates'] = dates
ndf['Counts_1'] = count
ndf=ndf.sort_values(by="Dates")
```

```python
#Send dates to CSV
ndf.Dates.to_csv("Update_Review/Sep/Dates/Random/" + OrgID + "_RC.csv",index=False, date_format='%m%d%Y')
```

```python
#Import results from search, calculate result length for each, merge with original data to compare.
fileIn = "Update_Review/Sep/Dates/Random/%s_C2.xlsx" %OrgID
Org2 = pd.read_excel(fileIn, parse_dates=True)
Org2 = Org2["Ord. Date"].value_counts()
Org2 = pd.DataFrame(Org2)
Org2 = Org2.reset_index()
Org2 = Org2.rename(columns={'index':'Dates', 'Ord. Date':'Counts_2'})
ndf.merge(Org2, how="outer", on="Dates")

```
